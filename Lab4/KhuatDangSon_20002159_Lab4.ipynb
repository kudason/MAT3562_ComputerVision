# Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Low Pass Frequency by using Distance smooth function

## Read Image and Convert Image to Gray

```python
img = cv2.imread('chess_board.png')
# Convert Image to gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(5, 5))
plt.imshow(gray, cmap='gray')
plt.show()
```

## Fourier Transformation

```python
# Fourier transform mation
f = np.fft.fft2(gray)
plt.imshow(np.log(np.abs(f)), cmap='gray')
plt.show()
```

## Frequency Shift

```python
shifted_f = np.fft.fftshift(f)
plt.imshow(np.log(np.abs(shifted_f)), cmap='gray')
plt.show()
```

## Low Pass Frequency

```python
low_pass = np.ones(shape=gray.shape)
plt.imshow(low_pass, cmap='gray', vmax=1, vmin=0)
```

```python
# Lay tam hinh tron
xc = gray.shape[1] // 2
yc = gray.shape[0] // 2

# Ban kinh (Do)
Do = 350

# Tinh toan khoang cach tu 1 diem den tam voi cong thuc: np.sqrt((x-xc)**2 + (y-yc)**2)
def d(x, y):
  return np.sqrt((x-xc)**2 + (y-yc)**2)

for x in range(gray.shape[1]):
  for y in range(gray.shape[0]):
    if d(x, y) > Do:
      low_pass[y,x] = 0

f = f * low_pass
plt.imshow(np.log(np.abs(low_pass)), cmap='gray')
plt.show()
```

## Invert Shift

```python
f = np.fft.ifftshift(f)
plt.imshow(np.log(np.abs(f)), cmap='gray')
plt.show()
```

## Invert Fourier Transform

```python
new_img = np.abs(np.fft.ifft2(f))
plt.imshow(new_img, cmap='gray')
plt.show()
```

# Fourier Transform

```python
for sigma in [0.1, 0.5, 1.0]:
  x = np.linspace(-5, 5, 201)
  y = np.exp(-x**2 / sigma)
  plt.plot(x, y, label = "".format(sigma))

plt.legend()
```

```python
for sigma in [0.1, 0.5, 1.0]:
  x = np.linspace(-5, 5, 201)
  y = 1 - np.exp(-x**2 / sigma)
  plt.plot(x, y, label = "".format(sigma))

plt.legend()
```

```python
# plot 3D function y = exp(-(x**2 + y**2) / sigma)
x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)

X, Y = np.meshgrid(x, y)

plt.figure(figsize=(10,8))
for i, sigma in enumerate([0.1, 0.5, 1.0, 5.0]):
  z = np.exp(-(X**2 + Y**2) / sigma)
  plt.subplot(2, 2, i+1)
  plt.imshow(z, cmap='gray')
  plt.title(f"sigma:{sigma}")

plt.show()
```

# Low pass frequency by using Gaussian smooth function

## Read Image and Convert to gray image

```python
img = cv2.imread('chess_board.png')
# Convert to gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Show image
plt.figure(figsize=(5, 5))
plt.imshow(gray, cmap='gray')
plt.show()
```

## Fourier Transformation

```python
f = np.fft.fft2(gray)
plt.figure(figsize=(5, 5))
plt.imshow(np.log(np.abs(f)), cmap='gray')
plt.show()
```

## Frequency Shift

```python
shifted_f = np.fft.fftshift(f)
plt.figure(figsize=(5,5))
plt.imshow(np.log(np.abs(shifted_f)), cmap='gray')
plt.show()
```

## Low pass frequency

```python
low_pass = np.ones(shape=gray.shape)

# Lay tam buc anh
xc = gray.shape[1] // 2
yc = gray.shape[0] // 2

# ban kinh
Do = 350

# Tinh khoang cach tu 1 diem den tam
def d(x, y):
  return np.sqrt((x-xc)**2 + (y-yc)**2)

for x in range(gray.shape[1]):
  for y in range(gray.shape[0]):
    low_pass[y,x] = np.exp(-d(x,y)**2 / (2*Do**2))

f = f * low_pass

plt.figure(figsize=(5,5))
plt.imshow(np.log(np.abs(low_pass)), cmap='gray')
plt.show()
```

## Invert Shift

```python
f = np.fft.ifftshift(f)
plt.figure(figsize=(5,5))
plt.imshow(np.log(np.abs(f)), cmap='gray')
plt.show()
```

## Invert Fourier Transform

```python
new_img = np.abs(np.fft.ifft2(f))
plt.figure(figsize=(5,5))
plt.imshow(new_img, cmap='gray')
plt.show()
```

# High pass frequency by using Gaussian smooth function

## Read Image and Convert to gray image

```python
img = cv2.imread('chess_board.png')
# Convert to gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Show image
plt.figure(figsize=(5, 5))
plt.imshow(gray, cmap='gray')
plt.show()
```

## Fourier Transformation

```python
f = np.fft.fft2(gray)
plt.figure(figsize=(5, 5))
plt.imshow(np.log(np.abs(f)), cmap='gray')
plt.show()
```

## Frequency shift

```python
shifted_f = np.fft.fftshift(f)
plt.figure(figsize=(5,5))
plt.imshow(np.log(np.abs(shifted_f)), cmap='gray')
plt.show()
```

## High pass frequency

```python
high_pass = np.ones(shape=gray.shape)

# Lay tam buc anh
xc = gray.shape[1] // 2
yc = gray.shape[0] // 2

# ban kinh
Do = 350

# Tinh khoang cach tu 1 diem den tam
def d(x, y):
  return np.sqrt((x-xc)**2 + (y-yc)**2)

for x in range(gray.shape[1]):
  for y in range(gray.shape[0]):
    high_pass[y,x] = 1 - np.exp(-d(x,y)**2 / (2*Do**2))

f = f * high_pass

plt.figure(figsize=(5,5))
plt.imshow(np.log(np.abs(high_pass)), cmap='gray')
plt.show()
```
